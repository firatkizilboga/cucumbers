```python
from sqlalchemy import create_engine

# Database connection details
MYSQL_USER = "admin"
MYSQL_PASSWORD = "31cuCUMbers"
MYSQL_HOST = "134.122.79.205"
MYSQL_PORT = 3306
MYSQL_DATABASE = "cucumdb"

# Create SQLAlchemy engine
engine = create_engine(
    f"mysql+pymysql://{MYSQL_USER}:{MYSQL_PASSWORD}@{MYSQL_HOST}:{MYSQL_PORT}/{MYSQL_DATABASE}"
)
```

```python
import os

nba_folder = "/Users/mertbozkurtlar/Documents/GitHub/cucumbers/raw_data/nba"
nba_files = [f for f in os.listdir(nba_folder) if f.endswith('.csv')]
nba_file_paths = [os.path.abspath(os.path.join(nba_folder, f)) for f in nba_files]
nba_file_paths
```

```python
for path in nba_file_paths:
    table_name = os.path.splitext(os.path.basename(path))[0]
    print("Uploading", table_name)
    table_name = table_name.replace(" ", "_")
    df = pd.read_csv(path)
    df.to_sql(
        table_name,
        engine,
        if_exists='replace',
        index=False
    )
    print("Uploaded", table_name)
```

```python
from sqlalchemy import text

with engine.connect() as connection:
    # 1. Get all tables from the current database
    tables_result = connection.execute(text("SHOW TABLES"))
    tables = [row[0] for row in tables_result.fetchall()]

    # 2. For each table, retrieve the CREATE TABLE statement and print it
    for table_name in tables:
        create_stmt_result = connection.execute(
            text(f"SHOW CREATE TABLE `{table_name}`")
        ).fetchone()
        
        # The second element in the returned tuple is the CREATE statement
        print(f"--- Schema for table: {table_name} ---")
        print(create_stmt_result[1])
        print()
```
